```python
def Calculate_depth(img, heightOfCamera, PointOfInterest, blindAreaRadius, f=50):
    blindAngle= atan2(heightOfCamera,blindAreaRadius)
    Beta= 180- blindAngle
    ImageCenterVertical= img.shape[0]/2
    Theta= atan2(f, PointOfInterest)- atan2(f,ImageCenterVertical)
    Alpha = 180-Beta-Theta
    D_actual= heightOfCamera*(1/tan(Beta)- 1/tan(Alpha))
    Total_distance= blindAreaRadius+D_actual
    return Total_distance
```

```python
dist= Calculate_depth(image,heightOfCamera, image.shape[0]-1900,blindAreaRadius)
print(dist)
```
